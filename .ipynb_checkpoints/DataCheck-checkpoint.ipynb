```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('data.csv')
```

```python
drop_cols = ['Отметка времени', 'Вопросы к авторам опроса', 
             'Ник в телеграм/Почта', 'Столбец 55', 'Столбец 44']
df = df.drop(columns=[col for col in drop_cols if col in df.columns])

likert_mapping = {
    'Не согласен': 1, 'Скорее не согласен': 2, 'Не уверен': 3,
    'Скорее согласен': 4, 'Согласен': 5
}

for col in df.select_dtypes(include=['object']).columns:
    if df[col].isin(likert_mapping.keys()).any():
        df[col] = df[col].map(likert_mapping)

df_encoded = pd.get_dummies(df, drop_first=True)
```

```python
df_encoded
```

```python
df_encoded.dtypes
```

```python
X = df_encoded.drop(['Средний балл по учебным дисциплинам за этот курс.'], axis = 1)
```

```python
y = df_encoded['Средний балл по учебным дисциплинам за этот курс.']
```

```python
from sklearn.feature_selection import r_regression

scores = r_regression(X, y)

feature_scores = pd.DataFrame({
    'feature': X.columns.tolist(),
    'correlation': scores
}).sort_values('correlation', key=abs, ascending=False)
negetive_features = feature_scores[feature_scores['correlation'].abs() < 0.2]

print(negetive_features['feature'].values)
```

```python
X = df_encoded.drop(negetive_features['feature'].tolist() + ['Средний балл по учебным дисциплинам за этот курс.'], axis = 1)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import KFold, GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
kf = KFold(n_splits=5, shuffle=True, random_state=42)

param_grid = {
    'n_estimators': [10, 50, 75, 100, 125, 150],
    'max_depth': [2, 3, 4, 5, 6, 7, 8]
}
```

```python
grid_search = GridSearchCV(
    estimator=RandomForestRegressor(random_state=42),
    param_grid = param_grid,
    scoring='neg_mean_squared_error',
    n_jobs=-1,
    verbose=1
)

grid_search.fit(X_train, y_train)
```

```python
forest_model = grid_search.best_estimator_
```

```python
y_pred = forest_model.predict(X_test)
```

```python
mean_squared_error(y_test, y_pred)
```

```python
r2_score(y_test, y_pred)
```

```python
mean_absolute_error(y_test, y_pred)
```

```python
from sklearn.inspection import permutation_importance

result = permutation_importance(
    grid_search, X_test, y_test, 
    n_repeats=10,     
    random_state=42, 
    n_jobs=-1          
)

```

```python
import matplotlib.pyplot as plt
import numpy as np

sorted_idx = result.importances_mean.argsort()

plt.figure(figsize=(6, 10))
plt.barh(range(len(sorted_idx)), result.importances_mean[sorted_idx], 
         xerr=result.importances_std[sorted_idx], align='center', color='skyblue')

plt.yticks(range(len(sorted_idx)), [col for col in X.columns])
plt.xlabel("Снижение точности (Permutation Importance)")
plt.title("Важность признаков на тестовой выборке")
plt.tight_layout()
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.savefig('artifacts/Permutation Importance.jpg')
plt.show()
```
